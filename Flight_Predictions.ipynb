```python
import seaborn as sns
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression

main_data = pd.read_csv("Flight_Info.csv")

#C: I was just going to use the Vistara since it is the most used airline in this dataset, but i decided to use the main_data df
vis_df = main_data[main_data["airline"] == "Vistara"]
vis_df.reset_index()

print(main_data.head)
```

```python
#C: Assigning the variables 
x = main_data[["duration", "days_left"]]
y = main_data[["price"]]
```

```python
#C: Training and testing the variables 
x_train, x_test, y_train, y_test = train_test_split(x, y, train_size = 0.8, test_size = 0.2, random_state=6)

lr = LinearRegression()

model = lr.fit(x_train, y_train)

y_predict= lr.predict(x_test)
```

```python
#C: This means this model unfortunately did not slay successful/ it did not work, as can be seen with the 5% result
print("Test score:")
print(str(round(lr.score(x_test, y_test)*100)) + "%")
```

```python
#C: Just to reveal how much of a poor connection there is betweeen the variables
plt.scatter(y_test, y_predict)

plt.xlabel("Prices")
plt.ylabel("Predicted Prices")
plt.show()
```
